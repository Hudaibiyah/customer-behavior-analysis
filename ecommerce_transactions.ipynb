Calculated Average Order value (AOV)

```python

aov = df['Purchase_Amount'].mean()
print(f"Average Order Value (AOV): {aov:.2f}")
```

```python
total_revenue = df['Purchase_Amount'].sum()
total_orders = len(df)
aov = total_revenue / total_orders
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Average Order Value: ${aov:.2f}")
# See AOV by Category
df.groupby('Product_Category')['Purchase_Amount'].mean().sort_values(ascending=False)
```

Repeat customer rate

```python
user_counts = df['User_Name'].value_counts()
print(user_counts)

repeat_customers = user_counts[user_counts > 1].count()
total_unique_customers = user_counts.count()

rcr = (repeat_customers / total_unique_customers) * 100

print(f"Repeat Customer Rate: {rcr:.2f}%")


```

Top products by popularity

```python
category_popularity = df['Product_Category'].value_counts()
print(category_popularity)
```

Top products by category

```python
category_revenue = df.groupby('Product_Category')['Purchase_Amount'].sum().sort_values(ascending=False)
print(category_revenue)
```

Revenue by country


```python
revenue_by_country = df.groupby('Country')['Purchase_Amount'].sum().sort_values(ascending=False)
print(revenue_by_country)



```

Customer lifetime value

```python
aov = df['Purchase_Amount'].mean()

purchase_frequency = len(df) / df['User_Name'].nunique()

repeat_rate = (df['User_Name'].value_counts() > 1).sum() / df['User_Name'].nunique()

churn_rate = 1 - repeat_rate
expected_lifespan_periods = 12

clv_lifespan = aov * purchase_frequency * expected_lifespan_periods
print(f"CLV Proxy (12 periods): ${clv_lifespan:,.2f}")

if churn_rate > 0:
    clv_churn = (aov * purchase_frequency) / churn_rate
    print(f"CLV (churn model): ${clv_churn:,.2f}")
else:
    print("CLV (churn model): Not reliable because churn rate is 0.")
```
